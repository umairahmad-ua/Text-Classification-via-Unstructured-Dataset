## Step 2: Pre-Process Data

```python


def deEmojify(inputString):
    return inputString.encode('ascii', 'ignore').decode('ascii')
def clean_tweets(tweet):
     #Remove URLs
    tweet = re.sub(r'^https?:\/\/.*[\r\n]*', '', tweet, flags=re.MULTILINE)
    tweet = re.sub(r'https?://.*?\\s+', '', tweet, flags=re.MULTILINE)
    pattern = re.compile('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+')
    tweet=pattern.sub('', tweet)
    #re.sub(r'(\s)http\w+', r'\1',tweet )
    #Remove long words
    longword = re.compile(r'\W*\b\w{15,500}\b')
    tweet=longword.sub('', tweet)
    #Remove short words
    shortword = re.compile(r'\W*\b\w{1,3}\b')
    tweet=shortword.sub('', tweet)
#remove emojis from tweet
    tweet = deEmojify(tweet)

    # lower capitalization
    tweet=tweet.lower()
    
    # Remove punctuation 
    remove = string.punctuation
    remove = remove.replace("@", "") # don't remove hyphens
    pattern = r"[{}]".format(remove) # create the pattern
    tweet=re.sub(pattern, "", tweet) 
    
    #Remove username
    tweet=re.sub(r'@.*?(?=\s)', '', tweet)
    #Remove digits
    tweet = ''.join([i for i in tweet if not i.isdigit()])
            
  
    return tweet
male_comments=clean_tweets(male_comments)
female_comments=clean_tweets(female_comments)
```

```python
#Remove URLs
#female_comments = re.sub(r"http\S+", "", female_comments)
#male_comments = re.sub(r"http\S+", "", male_comments)
#split
male_comments=male_comments.split("\n")
female_comments=female_comments.split("\n")
```

### Feature Extraction –  for stylometry based methods

```python
#1. Number of commas
commas_counter = combine_data.count(',')
print('commas_counter',commas_counter)
#2. Number of fullstop 
fullstop_counter = combine_data.count('.')
print('fullstop_counter',fullstop_counter)
#3. Number of @ 
At_counter = combine_data.count('@')
print('At_counter',At_counter)
#4. Number of ( 
leftbracket_counter = combine_data.count('(')
print('leftbracket_counter',leftbracket_counter)
#5. Number of ) 
rightbracket_counter = combine_data.count(')')
print('rightbracket_counter',rightbracket_counter)
#6. Number of ! 
exclamation_counter = combine_data.count('!')
print('exclamation_counter',exclamation_counter)
#7. Number of – 
minus_counter = combine_data.count('-')
print('minus_counter',minus_counter)
#8. Number of ?
qm_counter = combine_data.count('?')
print('qm_counter',qm_counter)
#9. Number of % 
per_counter = combine_data.count('%')
print('per_counter',per_counter)
#10. Number of & 
and_counter = combine_data.count('&')
print('and_counter',and_counter)
#11. Number of # 
hash_counter = combine_data.count('#')
print('hash_counter',hash_counter)
#12. Number of _ 
US_counter = combine_data.count('_')
print('US_counter',US_counter)
#13. Number of = 
equal_counter = combine_data.count('=')
print('equal_counter',equal_counter)
#14. Number of capital letters
capital_letters_counter=sum(1 for c in combine_data if c.isupper())
print('capital_letters_counter',capital_letters_counter)
#15. Number of semi colons 
sc_counter = combine_data.count(';')
print('sc_counter',sc_counter)
#16. Number of colons 
colons_counter = combine_data.count(':')
print('colons_counter',colons_counter)
#17. Number of spaces
space_counter = combine_data.count(' ')
print('space_counter',space_counter)
#18. Number of small characters 
small_letters_counter=sum(1 for c in combine_data if c.islower())
print('small_letters_counter',small_letters_counter)
#19. Number of forward slash (/) 
fs_counter = combine_data.count('/')
print('fs_counter',fs_counter)
#20. Number of digits 
digit_counter=sum(1 for c in combine_data if c.isdigit())
print('digit_counter',digit_counter)

 
```
